# üß† EDA on IBM HR Attrition Dataset

This notebook presents an Exploratory Data Analysis (EDA) of the IBM HR Analytics Employee Attrition dataset.
Our goal is to uncover the key factors that lead to employee attrition by analyzing various features including job role, environment satisfaction, work-life balance, and overtime status.


## üìö Importing Required Libraries

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## üìÅ Loading the Dataset

```python
# Load the dataset
data = pd.read_csv('Attrition.csv')
df_plot = data[['MaritalStatus', 'BusinessTravel', 'Department', 'Attrition']].copy()
```

## üîç Initial Data Exploration

```python
# Display the first few rows of the dataset
data.head()


# Display basic information about the dataset
data.info()

# Display summary statistics of the dataset
data.describe()
```

## üßπ Checking for Missing Values and Data Types

```python
# Data Preprocessing
# Check for missing values
missing_values = data.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
# Encode categorical variables
from sklearn.preprocessing import LabelEncoder

# Encode binary categorical variables
le = LabelEncoder()
data['Attrition'] = le.fit_transform(data['Attrition'])
data['Gender'] = le.fit_transform(data['Gender'])
data['OverTime'] = le.fit_transform(data['OverTime'])
data['Over18'] = le.fit_transform(data['Over18'])

# One-hot encode other categorical variables
data = pd.get_dummies(data, columns=['BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus'], drop_first=True)
```

## üîç Logistic Regression

```python
# Import necessary libraries for logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# Use the same dominant features for the model
dominant_features = ['Age', 'MonthlyIncome', 'YearsAtCompany', 'OverTime']
X_dom = data[dominant_features]
y = data['Attrition']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_dom, y, test_size=0.3, random_state=42)

# Create and train the logistic regression model
log_model = LogisticRegression(max_iter=1000, random_state=42)
log_model.fit(X_train, y_train)

# Predict the test set results
y_pred_log = log_model.predict(X_test)

# Calculate and print the accuracy score
accuracy = accuracy_score(y_test, y_pred_log)
print("Logistic Regression Test Accuracy:", accuracy)
```

## üîç RandomForest Classifier 

```python
# Import necessary libraries for modeling
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Select the dominant features for modeling
dominant_features = ['Age', 'MonthlyIncome', 'YearsAtCompany', 'OverTime']
X_dom = data[dominant_features]
y = data['Attrition']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_dom, y, test_size=0.3, random_state=42)

# Train a Random Forest classifier
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Evaluate the model on the test set
y_pred = model.predict(X_test)
print("Test Set Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))

# Define a function to predict attrition based on user input for dominant features
def predict_attrition(age, monthly_income, years_at_company, overtime):

    # Create a dataframe for the new input (must follow the same structure as X_dom)
    input_data = pd.DataFrame({
        'Age': [age],
        'MonthlyIncome': [monthly_income],
        'YearsAtCompany': [years_at_company],
        'OverTime': [overtime]
    })
    
    # Make a prediction (note: Attrition was label-encoded; 1 may indicate attrition and 0 no attrition)
    prediction = model.predict(input_data)[0]
    result = "Attrition" if prediction == 1 else "No Attrition"
    return result

# Example usage:
# You can replace these example values with any user inputs
example_age = 35
example_monthly_income = 5000
example_years_at_company = 5
example_overtime = 1  # 1 for Yes, 0 for No

prediction_result = predict_attrition(example_age, example_monthly_income, example_years_at_company, example_overtime)
print("Prediction for the input sample:", prediction_result)
```
